#### 2. Summary Statistics for Revenue Management:

* Calculate and display summary statistics (mean, median, standard deviation, etc.) for adr (average daily rate), lead_time, and total_stay (sum of stays_in_weekend_nights and stays_in_week_nights).


```python
hotel_data['total_stay'] = hotel_data["stays_in_week_nights"] + hotel_data["stays_in_weekend_nights"] #calculating total_stay
```

```python
hotel_data[['adr','lead_time','total_stay']].agg(["mean","median","std"]) #selecting the columns we want to aggregate
#afterwards we specify the aggregation functions which are to be run on the columns
```

```python
#alternative without agg
print(hotel_data[['adr','lead_time','total_stay']].mean())
print(hotel_data[['adr','lead_time','total_stay']].median())
print(hotel_data[['adr','lead_time','total_stay']].std())
```

```python
#alternative running on individual columns
hotel_data['adr'].agg(["mean","median","std"])
```

#### 3. Revenue Generation by Hotel Type:
* Calculate the total revenue generated (adr multiplied by total_stay) for each hotel type (Resort Hotel and City Hotel).
* Determine the average revenue per booking for each hotel type.

```python
hotel_data["revenue"] = hotel_data["adr"] * hotel_data["total_stay"] # creation of the revenue variable
hotel_data["revenue"]
```

```python
hotel_data[hotel_data["is_canceled"] == 0].groupby("hotel")["revenue"].sum() #grouping the filtered data frame and summing the revenue
```

```python
hotel_data[hotel_data["is_canceled"] == 0].groupby("hotel")["revenue"].mean()
```

```python
hotel_data[hotel_data["is_canceled"] == 0].groupby("hotel")["total_stay"].mean()
```

```python
hotel_data[hotel_data["is_canceled"] == 0].groupby("hotel")["adr"].mean()
```

#### 4. Revenue by Customer Segment:
* Group the data by market_segment and calculate the total and average revenue for each segment.
* Identify which market segment generates the most revenue on average.




```python
hotel_data[hotel_data["is_canceled"] == 0].groupby("market_segment")["revenue"].agg(["sum","mean"])
```

```python
hotel_data[hotel_data["is_canceled"] == 0].groupby("market_segment")["revenue"].mean().sort_values(ascending=False)
```

#### 5. Impact of Lead Time on Revenue:
* Calculate the average adr for bookings with different lead times (e.g., short lead time < 30 days, medium lead time 30-90 days, long lead time > 90 days).
* Analyze how lead time affects the revenue.



```python
def getLeadtimeCategory(leadtime):
    if leadtime < 30:
        return("short")
    elif leadtime <= 90:
        return("medium")
    else:
        return("long")
    
hotel_data["leadtime_category"] = hotel_data["lead_time"].apply(getLeadtimeCategory)

hotel_data[["lead_time","leadtime_category"]]
```

```python
hotel_data[hotel_data["is_canceled"] == 0].groupby(["hotel","leadtime_category"])["adr"].mean()
```

```python
hotel_data[hotel_data["is_canceled"] == 0].groupby("leadtime_category")["revenue"].mean()
```

#### 6. Cancellation Impact on Revenue:
* Calculate the total potential revenue lost due to cancellations.
* Determine the average adr and lead_time for canceled bookings versus non-canceled bookings.



```python
hotel_data[hotel_data["is_canceled"] == 1]['revenue'].sum()
```

```python
hotel_data.groupby(["hotel","is_canceled"])[["adr","lead_time"]].mean()
```

#### 7. Revenue by Room Type:
* Group the data by assigned_room_type and calculate the average adr and total revenue for each room type.
* Identify the most and least profitable room types.



```python
hotel_data[hotel_data["is_canceled"] == 0].groupby("assigned_room_type")["adr"].mean().sort_values()
```

```python
hotel_data[hotel_data["is_canceled"] == 0].groupby("assigned_room_type")["revenue"].sum().sort_values()
```

#### 8. Length of Stay and Revenue:
* Analyze the relationship between total_stay (length of stay) and adr.
* Determine the average revenue per booking for different lengths of stay (e.g., short stay < 3 nights, medium stay 3-7 nights, long stay > 7 nights).



```python
hotel_data[hotel_data["is_canceled"] == 0][["total_stay","adr"]].corr()
```

```python
def getTotalstayCategory(totalstay):
    if totalstay < 3:
        return("short")
    elif totalstay <= 7:
        return("medium")
    else:
        return("long")

hotel_data["totalstay_category"] = hotel_data["total_stay"].apply(getTotalstayCategory)
```

```python
hotel_data.groupby("totalstay_category")["revenue"].mean()
```

#### 9. Booking Channels and Revenue:
* Group the data by distribution_channel and calculate the total and average revenue for each channel.
* Identify which distribution channel is the most effective in terms of revenue generation.



```python
hotel_data.groupby("distribution_channel")["revenue"].agg(["mean","sum"]).sort_values(ascending=False, by = "sum")
```

#### 10. Seasonality and Revenue:
* Calculate the total and average revenue for each month.
* Identify peak and off-peak months in terms of revenue generation.

```python
hotel_data.groupby(["hotel","arrival_date_month"])["revenue"].agg(["sum"]).sort_values(ascending=False, by = ["hotel","sum"])
```
